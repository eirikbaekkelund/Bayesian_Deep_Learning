```python
import numpy as np
import torch
from torch import autograd
import torch.distributions as dist

import matplotlib.pyplot as plt
```

## Classic autograd usage

To declare a torch tensor as a leaf node in a computation graph, use `requires_grad`. Notice the difference:

```python
x = torch.FloatTensor([2.2, 1.1])
x
```

```python
x.requires_grad_(True)
```

A tensor that has `requires_grad=True` has a `.grad` attribute. This is initialized to `None`.

```python
x.grad is None
```

You can see the computation graph being built dynamically! For internal nodes, it shows up as `grad_fn`.

```python
y = x.pow(2)
y
```

```python
z = (torch.sin(x) - y).sum()
z
```

Once you've done some computation, you can call `.backward()` on a scalar output to accumulate gradients in the leaf nodes of the graph:

```python
z.backward()
x.grad
```

Calling `z.backward()` again will raise an error. If you want second derivatives, you'll need to explicitly tell it to "retain" the graph.

Also, be careful if you now "reuse" this value of `x`!

```python
(x + 2).sum().backward()
x.grad
```

This value is wrong, because when `.backward()` is called, it does not reset the currently stored `.grad` attribute to zero! Instead, it accumulates more partial derivatives. You can manually zero it out though, to get the expected result. Note that `.zero_()`, and all the methods with underscores, are generally "in-place" operations. You will usually want to avoid these, but in this case it's actually what we want.

```python
x.grad.zero_()
(x + 2).sum().backward()
x.grad
```

## Torch optimizers

There are many optimization tools built into pytorch. Here we will show how to use the `torch.optim.SGD` class to minimize a test function.

```python
def fn(x):
    return 0.1*x.pow(2) - x + x.sin()

domain = torch.linspace(-15, 15, 1000)
plt.plot(domain, fn(domain));
```

Most of the time you use gradients in pytorch, particularly for deep learning models, it will be something like the following incantation:

```python
# set initial point
x = torch.tensor(-10.0, requires_grad=True)

# define optimizer
opt = torch.optim.SGD((x, ), lr=0.1, momentum=0.8)

# iterate
N_steps = 30
# track progress
values = []
for _ in range(N_steps):
    # zero the gradients
    opt.zero_grad()
    # forward pass
    y = fn(x)
    # backward pass
    y.backward()
    # save value for reference (`.item()` returns the raw scalar!)
    values.append((x.item(), y.item()))
    # do one optmizer step
    opt.step()
```

Plot what happened:

```python
plt.plot(domain, fn(domain));
plt.scatter(torch.FloatTensor(values)[:,0], 
            torch.FloatTensor(values)[:,1], 
            c=torch.linspace(0, 1, len(values)), 
            cmap='plasma_r');
```

## Plotting whole derivatives

This isn't a very common use-case, but you can do it!

```python
inputs = torch.linspace(-15, 15, 100).requires_grad_(True)
outputs = fn(inputs)

# this is one way to compute the gradient in parallel at each point,
# by passing in `ones` as the starting values everywhere
g = autograd.grad(outputs, inputs, grad_outputs=torch.ones_like(outputs), create_graph=True)

# this is how you can compute a second derivative
gg = autograd.grad(g[0], inputs, grad_outputs=torch.ones_like(outputs))

# When you are doing something that doesn't require the gradients,
# like plotting, then you should wrap your computation in
# `torch.no_grad()`.
with torch.no_grad():
    plt.plot(inputs, outputs)
    plt.plot(inputs, g[0])
    plt.plot(inputs, gg[0])
    plt.legend(['function', '1st derivative', '2nd derivative'])
```

### Note: variable which require gradients can't be cast to numpy

They can be `.detach()`d, or wrap the block of code in `torch.no_grad()`. Of course, this will create a "break" in the computation graph!

```python
outputs.numpy()
```

```python
with torch.no_grad():
    print(outputs.numpy())
```

## Comparing stochastic gradient estimators

Suppose we are interested we have a distribution $p(x|a) = \mathcal{N}(x | a, 1)$ and a function $f(x) = x^2 - 2$.

We're interested in estimating $g = \nabla_a \mathbb{E}_{p(x | a)}[f(x)]$.

There are two Monte Carlo estimators we discussed (see notes). The first one, the score function estimator, can be estimated using samples $x_1,\dots,x_N \sim p(x|a)$:

$$\hat g^{S} = \frac{1}{N} \sum_{i=1}^N f(x_i)\nabla_a \log p(x_i|a).$$

The second is the reparameterized estimator, which takes samples $\epsilon_1,\dots, \epsilon_N \sim \mathcal{N}(0, 1)$, from which we can compute values $x_i = a + \epsilon_i$. This means that the value $x_i$ is differentiable w.r.t. $a$, and we approximate the gradient as

$$\hat g^{R} = \frac{1}{N} \sum_{i=1}^N \nabla_a f(x_i).$$

Typically, the reparameterized gradient estimator will have much lower variance for the same number of samples $N$.


```python
def fn(x):
    return x**2 - 2

a = torch.tensor(0.1, requires_grad=True) # arbitrary initial value
p = dist.Normal(a, 1)
```

```python
def score_fn_est(N=1):
    x = p.sample((N,))
    grad_p = autograd.grad((fn(x) * p.log_prob(x)).mean(), a)[0]
    return grad_p
```

```python
sf_est = [score_fn_est().item() for _ in range(1000)]
print("mean =", np.mean(sf_est), "variance =", np.var(sf_est))
plt.hist(sf_est, bins=20);
```

```python
def rep_est(N=1):
    x = p.rsample((N,))
    grad_p = autograd.grad(fn(x).mean(), a)[0]
    return grad_p
```

```python
rp = [rep_est().item() for _ in range(1000)]
print("mean =", np.mean(rp), "variance =", np.var(rp))
plt.hist(rp, bins=20);
```

```python
plt.boxplot([sf_est, rp]);
plt.xticks((1,2),["score function", "reparameterized"]);
```
